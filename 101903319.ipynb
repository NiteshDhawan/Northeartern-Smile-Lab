```python
import tensorflow as tf
from keras.models import Sequential
from keras.layers import Activation, Dropout, Flatten, Dense, Conv2D, MaxPool2D
from keras.preprocessing.image import ImageDataGenerator
```

```python
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)

# Input data files are available in the read-only "../input/" directory
# For example, running this (by clicking run or pressing Shift+Enter) will list all files under the input directory

import os
for dirname, _, filenames in os.walk('/kaggle/input'):
    for filename in filenames:
        print(os.path.join(dirname, filename))
```

```python
model=Sequential()
```

```python
model.add(Conv2D(filters=32, kernel_size=3, input_shape=[128,128,3], activation='relu', kernel_regularizer='l1'))
model.add(MaxPool2D((2,2)))
model.add(Conv2D(filters=64, kernel_size=3, input_shape=[128,128,3], activation='relu', kernel_regularizer='l1'))
model.add(MaxPool2D((2,2)))
model.add(Conv2D(filters=128, kernel_size=3, input_shape=[128,128,3], activation='relu', kernel_regularizer='l1'))
model.add(MaxPool2D((2,2)))
model.add(Conv2D(filters=256, kernel_size=3, input_shape=[128,128,3], activation='relu', kernel_regularizer='l1'))
model.add(MaxPool2D((2,2)))
```

```python
model.add(Flatten())
```

```python
model.add(Dense(10, activation='relu'))
```
